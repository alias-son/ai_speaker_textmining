### [ 필요 모듈 import ]

```python
import requests                       # html 파일을 받아오기 위함
from bs4 import BeautifulSoup         # BeautifulSoup!
import re                             # 정규표현식 사용을 위함
import datetime                       # 포스트 publishing 날짜 문자열을 datetime으로 변환하기 위함
```

<br/><br/>
### [ 블로그 URL을 통해 html 코드 가져오기 ]
- 참고로 네이버 블로그는 원본 HTML 소스코드를 얻기 위해 URL을 전처리하는 작업이 한번 더 필요하다

```python
base_url = "http://blog.naver.com"
url = "https://blog.naver.com/victoryace89/221482949629"
#url = "https://blog.naver.com/corn38/221458539212"
```

```python
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.text.encode("utf-8"), "html.parser")
```

```python
soup
```

```python
additional_url = soup.find(id='mainFrame').get('src')
```

```python
additional_url
```

```python
url2 = base_url + additional_url
```

```python
url2
```

```python
r = requests.get(url2)
soup = BeautifulSoup(r.text.encode("utf-8"), "html.parser")
soup
```

<br/><br/>
### [ 블로그 포스트 제목 추출 ]

```python
soup.find('meta', attrs={'property':'og:title'}).get('content')
```

<br/><br/>
### [ 블로그 작성일 추출 ]

```python
import datetime
date_str = soup.find('span', attrs={'class':'se_publishDate pcol2'}).get_text()
publish_datetime = datetime.datetime.strptime(date_str, '%Y. %m. %d. %H:%M')
print(publish_datetime)
```

<br/><br/>

```python
soup.find('meta', attrs={'property':'naverblog:nickname'}).get('content')
```

### [ 블로그 본문 추출 ]

```python
import re
```

```python
temp = soup.findAll("span", attrs= {'id': re.compile('^SE')})
temp
```

```python
texts = []
for t in temp:
    temp_text = t.get_text()
    if temp_text != '\u200b':
        texts.append(temp_text)
```

```python
texts
```

```python
print("\n".join(texts))
```
