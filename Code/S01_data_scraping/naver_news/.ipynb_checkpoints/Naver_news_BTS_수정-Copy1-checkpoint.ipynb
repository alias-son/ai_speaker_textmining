```python
from bs4 import BeautifulSoup
import requests
```

```python
import time
import random
from tqdm import tqdm_notebook
```

```python
import re
import pandas as pd
```

## Naver news에서 title, publisher, date, url 가져오기

```python
def get_articles(url):
    
    titles, naver_urls, pubs, dates = [], [], [], []
    
    r = requests.get(url)
    soup = BeautifulSoup(r.text.encode("utf-8"), "html.parser")
    lis = soup.find_all('li', attrs={'id':re.compile(r'sp_nws')})
    
    for li in lis:

        title = li.find('a', attrs={'class':re.compile(r'_sp_each_title')}).get('title').strip()
        if li.find('a', attrs={'class':re.compile(r'_sp_each_url')}).text:
            naver_url = li.find('a', attrs={'class':re.compile(r'_sp_each_url')}).get('href').strip()
        pub = li.find('span', attrs={'class':'_sp_each_source'}).text.strip()
        date = li.find('span', attrs={'class':'bar'}).next_sibling.strip()

        titles.append(title)
        naver_urls.append(naver_url)
        pubs.append(pub)
        dates.append(date)
        print(title, naver_url, pub, date)
    
    return titles, naver_urls, pubs, dates
```

```python
date_pd = pd.date_range('2017-09-01', '2019-04-30').strftime("%Y%m%d") # 네이버 클로바 출시일- 2017.10
date_pd[:10]
```

```python
q = '"카카오%20미니"%20스피커'

total_titles, total_urls, total_pubs, total_dates = [], [], [], []

for date in tqdm_notebook(date_pd):
    for page in range(1,100):
        print(page)
        try: 
            time.sleep(0.1)

            url = 'https://search.naver.com/search.naver?&where=news&query='+q+'&sm=tab_pge&sort=0&docid=&nso=so:r,p:from'+str(date)+'to'+str(date)+',a:all&start='+str(page)

            title, naver_url, pub, date = get_articles(url)

            if naver_url[0] in total_urls:# 페이지 끝나면 멈춤
                break

            total_titles += title
            total_urls += naver_url
            total_pubs += pub
            total_dates += date

        except:
            break
```

```python
len(total_titles), len(total_pubs), len(total_dates), len(total_urls)
```

```python
naver_news = pd.DataFrame(data = {'title': total_titles, 'publisher': total_pubs, 'date':total_dates, 'url': total_urls})
naver_news.head()
```

```python
texts = []
for url in tqdm_notebook(naver_news['url']):
    if 'news.naver.com/main' in url:
        print(url)
        
        r = requests.get(url)
        soup = BeautifulSoup(r.text, 'lxml')
        
        try:
            text = soup.find('div', attrs = {'id': 'articleBody'}).text
            texts.append(text.split('이 기사는')[0].split('}')[1].strip())
        except:
            text = soup.find('div','end_body_wrp').text
            texts.append(text.split('▶')[0].strip())
    
    else:
        texts.append('NA')
```
