# preprocessing

```python
pos = []
with open('pos_texts.txt', 'r', encoding='utf-8') as p:
    for i in p.readlines():
        pos.append(i.split('\n')[0])
len(pos)
```

```python
import nltk.tokenize
sentences = list(map(nltk.tokenize.sent_tokenize, pos))
sentences[:5]
```

```python
words = [list(map(nltk.tokenize.word_tokenize, sentence)) for sentence in sentences]
words[:5]
```
