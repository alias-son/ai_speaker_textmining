## CLOVA 뉴스

```python
import pandas as pd
```

```python
clova_df = pd.read_csv('data/clova_speaker_news.csv', index_col=0)
clova_df = clova_df.sort_values(['date'],ascending=True).reset_index(drop=True) # 날짜로 정렬
```

```python
clova_df.head()
```

```python
clova_df.info()
```

```python
clova_df.isnull().sum()
```

```python
clova_NullData_texts = clova_df.loc[clova_df.loc[:,'texts'].isna(),:]
clova_NullData_texts
```

```python
clova_NullData_texts.date.unique()
```

```python
clova_NullData_texts.publisher.unique()
```

```python
clova_NullData_date = clova_df.loc[clova_df.loc[:,'date'].isna(),:]
clova_NullData_date
```

#### null 값 및 중복 기사 제거

```python
clova_df = clova_df.dropna(subset = ['texts','date'], axis=0).drop_duplicates(subset = ['texts'])
```

```python
clova_df.info() # 659개 기사
```

```python
clova_df.to_csv("data/nonull_clova_news.csv", encoding='utf8')
```
