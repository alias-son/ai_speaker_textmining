```python
def txt_to_df(file):
    import pandas as pd
    
    file = '/Users/yaelinjo/Desktop/notebook/Textmining/mallet/output/'+ file
    with open(file, encoding='utf8')as f:
        
        lines = f.readlines()
        
        topic0 = {}
        for line in lines[1:22]:
            if '<default>' in line:
                default = float(line.split(' ')[2].rstrip('\n'))
            else:
                topic0[line.split(' ')[1]] = float(line.split(' ')[2].rstrip('\n'))
        
        topic1 = {}
        for line in lines[23:44]:
            if '<default>' in line:
                default = float(line.split(' ')[2].rstrip('\n'))
            else:
                topic1[line.split(' ')[1]] = float(line.split(' ')[2].rstrip('\n'))
        
        topic2 = {}
        for line in lines[45:66]:
            if '<default>' in line:
                default = float(line.split(' ')[2].rstrip('\n'))
            else:
                topic2[line.split(' ')[1]] = float(line.split(' ')[2].rstrip('\n'))
        
        topic3 = {}        
        for line in lines[67:88]:
            if '<default>' in line:
                default = float(line.split(' ')[2].rstrip('\n'))
            else:
                topic3[line.split(' ')[1]] = float(line.split(' ')[2].rstrip('\n'))
        
        topic4 = {}  
        for line in lines[89:110]:
            if '<default>' in line:
                default = float(line.split(' ')[2].rstrip('\n'))
            else:
                topic4[line.split(' ')[1]] = float(line.split(' ')[2].rstrip('\n'))
        
        topic5 = {}  
        for line in lines[111:132]:
            if '<default>' in line:
                default = float(line.split(' ')[2].rstrip('\n'))
            else:
                topic5[line.split(' ')[1]] = float(line.split(' ')[2].rstrip('\n'))
        
        topic6 = {}  
        for line in lines[133:154]:
            if '<default>' in line:
                default = float(line.split(' ')[2].rstrip('\n'))
            else:
                topic6[line.split(' ')[1]] = float(line.split(' ')[2].rstrip('\n'))
        
        topic7 = {}  
        for line in lines[155:176]:
            if '<default>' in line:
                default = float(line.split(' ')[2].rstrip('\n'))
            else:
                topic7[line.split(' ')[1]] = float(line.split(' ')[2].rstrip('\n'))
        
        topic8 = {}  
        for line in lines[177:198]:
            if '<default>' in line:
                default = float(line.split(' ')[2].rstrip('\n'))
            else:
                topic8[line.split(' ')[1]] = float(line.split(' ')[2].rstrip('\n'))
        
        topic9 = {}  
        for line in lines[199:220]:
            if '<default>' in line:
                default = float(line.split(' ')[2].rstrip('\n'))
            else:
                topic9[line.split(' ')[1]] = float(line.split(' ')[2].rstrip('\n'))
                
                
        topic_parameter = pd.DataFrame([topic0, topic1, topic3, topic3, topic4, topic5,
                                        topic6, topic7, topic8, topic9],
                                       index = ['topic0', 'topic1', 'topic2', 'topic3', 'topic4', 'topic5', 
                                       'topic6', 'topic7', 'topic8', 'topic9'])
        
        
    return topic_parameter
```

```python
topic_df = txt_to_df('topic_parameter.txt')
topic_df
```

```python
topic_df.index.set_names('topics', inplace=True)
topic_df.columns.set_names('month', inplace=True)
topic_df
```
