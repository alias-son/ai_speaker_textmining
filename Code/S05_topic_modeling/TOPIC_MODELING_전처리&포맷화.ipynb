## CLOVA

```python
import pandas as pd
from tqdm import tqdm_notebook
import re
```

```python
clova_df = pd.read_csv('data/nonull_clova_news.csv',index_col=0)
clova_df # 659개
```

```python
clova_news = list(clova_df.texts)
```

#### 미등록단어 추출

```python
from kornounextractor.noun_extractor import extract
```

```python
unregistered_word = []

for news in tqdm_notebook(clova_news):
    unregistered_word.append(extract(news))
```

```python
unregistered_word = sum(unregistered_word, [])
unregistered_word = set(unregistered_word)
```

```python
for word in unregistered_word:
    with open("clova_unregistered.txt", "a", encoding="utf8") as f:
        f.write("{}\tNNG\n".format(word))
```

#### PoS tagging

```python
from konlpy.tag import Komoran
```

```python
komoran = Komoran(userdic="clova_unregistered.txt")
```

```python
pos_tagging_list = list(map(komoran.pos, clova_news))
```

```python
clova_pos_tagging_list = pos_tagging_list
```

```python
import re

p = re.compile('NN.*|SL')
```

```python
clova_docs = []
for doc in clova_pos_tagging_list:
    want_words = []
    for word, pos in doc:
        if p.match(pos):
            want_words.append(word)
    clova_docs.append(want_words)
```

#### 한 음절 단어 제거

```python
for doc in tqdm_notebook(clova_docs):
    for word in doc:
        if len(word) == 1:
            while word in doc:
                doc.remove(word)
```

```python
import pickle
with open('data/clova_docs_불용어전처리전', 'wb') as fp:
    pickle.dump(clova_docs, fp)

#with open('data/clova_docs_불용어전처리전', 'rb') as fp:
#clova_docs = pickle.load(fp)
```

#### 불용어 제거

```python
with open('data/clova_docs_불용어전처리전', 'rb') as fp:
    clova_docs = pickle.load(fp)
```

```python
stopwords = [line.strip() for line in open("data/stopwordsKor.txt", encoding="utf8")]
```

```python
clova_final_docs = []
for doc in clova_docs:
    unique_NN_words = set(doc)
    final_NN_words = doc
    
    for word in unique_NN_words:
        if word in stopwords:
            while word in final_NN_words: final_NN_words.remove(word)
    clova_final_docs.append(final_NN_words)
```

#### document 별 단어를 각각의 document로 바꾸기

```python
clova_documents = []
for doc_words in clova_final_docs:
    clova_document = " ".join(doc_words)
    clova_documents.append(clova_document)
```

#### mallet input txt file 생성

```python
title = list(clova_df['title'])
date = list(clova_df['date'])
```

```python
with open("data/clova_news_input_data.txt", "w", encoding="utf8") as f:
    for i in range(0,len(clova_documents)):
        f.write("{}\t{}\t{}\n".format(title[i],
                                      date[i][0:7],
                                      clova_documents[i]))
```
