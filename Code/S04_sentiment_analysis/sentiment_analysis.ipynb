# TEST 1. 네이버 클로바 블로그 데이터 (new)

```python
import pandas as pd
```

```python
# CSV 데이터 불러오기
blog_df = pd.read_csv('../../data/naver_blog/naverclova_speaker_blog.csv', encoding='utf-16', index_col=0)
```

```python
blog_df.head()
```

```python
blog_df.text[:5]
```

```python
# 블로그 본문 컬럼이 Not null인 총 개수 확인
len(blog_df[blog_df.text.notnull()])
```

```python
# 블로그 본문 컬럼이 Not null인 row를 제외하기
blog_df = blog_df[blog_df.text.notnull()].copy()
```

```python
# 블로그 본문에 있는 개행문자(\n, \r)를 제거
blog_df.text = blog_df.text.map(lambda x:x.replace("\n", " ").replace("\r", " ").replace('\xa0', ""))
```

<br/><br/>

### Tokenize 하기

#### 1. 문장 단위로 tokenize

```python
import nltk.tokenize
sentences = list(map(nltk.tokenize.sent_tokenize, blog_df.text))
sentences[:5]
```

```python
from konlpy.tag import Komoran

komoran = Komoran(userdic='./unregistered.txt')
#pos_tagging_list = []
#for doc in sentences:
#    pos_tagging_list.append(list(map(komoran.pos, doc)))
pos_tagging_list = [list(map(komoran.pos, doc)) for doc in sentences]
```

```python
pos_tagging_list[:1]
```

```python
docs = []
for doc in pos_tagging_list:
    sentences = []
    for sent in doc:
        want_words = []
        for word, pos in sent:
            want_words.append(word)
        sentences.append(want_words)
    docs.append(sentences)
```

```python
docs[:1]
```

<br/><br/>
### 감성분석 코드 

```python
def readDictionary(file):
    fh = open(file, mode="r", encoding="utf-8")
    dict_list = {}
    for line in fh:
        fields = line.split(',')
        n_token = ''
        toks = fields[0].split(";")
        for tok in toks:
            n_token += tok.split("/")[0]
        #print("element " + n_token + " -- " + fields[len(fields)-2] + " : " + fields[len(fields)-1])
        sent_dict = {}
        sent_dict['word'] = n_token
        sent_dict['polarity'] = fields[len(fields)-2]
        sent_dict['score'] = fields[len(fields)-1].replace("\n","")
        dict_list[n_token] = sent_dict
    return dict_list

if __name__ == '__main__':
    file_name = './polarity.csv'
    dict_list = readDictionary(file_name)
    
    doc_list = []
    sentence_list = []
    sentence_idx_list = []
    avg_list = []
    
    for d_idx, doc in enumerate(docs):

        for s_idx, sent in enumerate(doc):
            total_score = 0.0
            count = 0
            for _str in sent:
                if len(_str) > 0:
                    score = 0.0
                    dictionary_ele = dict_list.get(_str)
                    if (dictionary_ele != None):
                        polarity = dictionary_ele.get('polarity')
                        score = dictionary_ele.get('score')
                        if (polarity == 'NEG'):
                            score = -float(score)
                            #print("(-)부정어:", _str)
                            count += 1
                        else:
                            score = float(score)
                            #print("(+)긍정어:", _str)
                            count += 1
                        #print(_str + " == " + polarity + " " + str(score))
                        total_score += score
                    else:
                        total_score += score

            if (count != 0):
                avg_score = total_score/count
                #print("AVG SCORE " + str(avg_score))
            doc_list.append(d_idx)
            sentence_list.append(" ".join(sent))
            sentence_idx_list.append(s_idx)
            avg_list.append(avg_score)
```

```python
len(dict_list)
```

```python
dict_list
```

```python
dict_list.keys()
```

<br/><br/>

### 결과 확인
- 아... ㅠㅠ

```python
sentiment_df = pd.DataFrame(data={'document_idx': doc_list, 'sentence_idx': sentence_idx_list, 
                                  'sentence_text': sentence_list, 'sentiment_score': avg_list})
```

```python
sentiment_df
```

```python
temp = sentiment_df[sentiment_df['sentiment_score'] == sentiment_df['sentiment_score'].min()]
for text, score in zip(temp['sentence_text'], temp['sentiment_score']):
    print("{} : {}".format(score, repr(text)))
```

```python
temp = sentiment_df[sentiment_df['sentiment_score'] < 0]
for text, score in zip(temp['sentence_text'], temp['sentiment_score']):
    print("{} : {}\n".format(score, repr(text)))
```
